```python
# import the neccessary modules
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Load and import the dataset
df = pd.read_csv('energydata_complete.csv')
```

```python
# check out the columns and information of the datasetdf.head()
df.info()
df.describe()
```

```python
# Remove the specified columns: "date", "lights", and keep "Appliances" as the target variable
df = df.drop(columns=["date", "lights"])

# Split the dataset into features (X) and target (y)
X = df.drop(columns=["Appliances"])
y = df["Appliances"]

# Split the data into a 70-30 train-test set with a random state for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the MinMaxScaler and fit_transform it on the training data, then transform the test data
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize and fit a multiple linear regression model on the training data
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Obtain predictions on the test set
y_pred = model.predict(X_test_scaled)

# Calculate the Root Mean Squared Error (RMSE) on the test set
rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f" (RMSE): {rmse:.3f}")
```
